```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
np.random.seed(10)
sample_data = np.random.exponential(scale=2.0, size=1000)
```

```python
print(sample_data)
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
sns.histplot(sample_data,bins=6,kde=True,color="skyblue")
plt.title("histogram")
plt.xlabel("score")
plt.ylabel("frequency")

plt.subplot(1,2,2)
sns.boxplot(x=sample_data,color="lightgreen")
plt.title("boxplot")
plt.xlabel("score")

plt.tight_layout()
plt.show
```
