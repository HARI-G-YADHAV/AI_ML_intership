```python
import seaborn as sns

iris = sns.load_dataset("iris")
```

```python
print(iris.head())
```

```python
print(iris['species'].unique())
```

```python
sns.set_theme(style="whitegrid", palette = "deep")

plot = sns.barplot(x="species",y="sepal_length",data = iris,palette = "pastel")

plot.set_xlabel("iris species name")
plot.set_ylabel("iris sepal length")
plot.set_title("bar plot of species vs sepallength")

plot.figure.tight_layout()
plot.figure.show()
```
